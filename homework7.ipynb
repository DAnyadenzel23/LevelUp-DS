13.Посчтитайте максимальный и минимальный уровень холестерина для каждого типа chest_pain_type, rest_ecg, thalassemia. 

```python
attributes = ['chest_pain_type', 'rest_ecg', 'thalassemia']
for att in attributes:
    for sign in df[att].unique():
        print(f'{df[df[att] == sign]['cholesterol'].max()} - максимальный уровень холестрина для признака:подгруппы {att}:{sign}')
        print(f'{df[df[att] == sign]['cholesterol'].min()} - минимальный уровень холестрина для признака:подгруппы {att}:{sign}')
        print('---------------------')
```

15.У какого категориального признака наблюдается самый сильный дизбаланс классов?

```python
for col in cols:
    print(f'{df[col].value_counts(normalize=True).min()} - минимальная доля класса признака {col}')
```

```python
lst_of_min_represent = {col: df[col].value_counts(normalize=True).min() for col in cols}
print(f'{min(lst_of_min_represent, key=lst_of_min_represent.get)} - признак с максимальным дисбалансом')
```
