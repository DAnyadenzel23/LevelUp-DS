Задание 9
Напишите функцию, которая на вход принимает 2 вектора и возвращает тот из них, что больше. Функцию необходимо векторизовать.

```python
#Функция1(без векторизации) - с итерацией по векторам через цикл for и подсчет количества бОльших координат одного индекса...на основании этого
#           возвращает тот вектор, что больше
def greater_v(v_1, v_2):
    if v_1.shape[1] == v_2.shape[1]:
        counter_1 = 0
        counter_2 = 0
        r = v_2.shape[1]
        for i in range(r):
            if v_1[0][i] > v_2[0][i]:
                counter_1 += 1
            else:
                counter_2 += 1
        if counter_1 > counter_2:
            return v_1
        else: 
            return v_2
    return 'Вектора имеют разную длину'        
    
  
```

```python
a = np.array([[1, 4, 8]])
b = np.array([[5, 1, 9, 88]])
greater_v(a,b)
```

```python
a = np.array([[1, 4, 8]])
b = np.array([[5, 1, 9]])
greater_v(a,b)
```

```python
# Функция2(без векторизации) - сразу настроена на работу с векторами, использует функцинал numpy, а именно метод np.greater, который
#                              поэлементно сравнивает два массива. Считается ли это векторизованной функцией????
def max_v(v_1, v_2):
    try:
        res_arr = np.greater(v_1, v_2)
        if np.sum(res_arr) > res_arr.shape[1]/2:
            return v_1
        else: 
            return v_2
    except ValueError:
        print('Вектора имеют разную длину')
            
max_vext = np.vectorize(max_v)
max_v(a,b)
```

```python
a = np.array([[1, 4, 8, 10]])
b = np.array([[5, 1, 9]])
max_v(a, b)
```

```python
# Функция3(для скалярных вычислений) - 1)для последующей ее векторизации, для последующей подачи в нее векторов.
#                                      2)не может вернуть один из исходных векторов, который по своим координатам окажется больше.
#                                      3)внутри нельзя писать никакой логики для проверки длины векторов, так как векторизованная функция 
#                                      будет подавать в отдельности каждый разный сразу пары координат векторов для последующего их сравнения.
def max_value(x,y):
    if x > y:
        return x
    else:
        return y
   
    
    
    
vector_max_value = np.vectorize(max_value)

a = np.array([[1, 4, 8,]])
b = np.array([[5, 4, 9]])
vector_max_value(a, b)
```
